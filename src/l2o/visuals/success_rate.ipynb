```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
optimization_success_rates = np.array([
    [0.95, 0.92, 0.85],  # Method 1 success rates for Test 1, Test 2, Test 3
])

objective_success_rates = np.array([
    [0.65, 0.45, 0.35],  # Method 2 success rates for Test 1, Test 2, Test 3
])

num_tests = len(optimization_success_rates[0])
num_methods = 2  # Number of methods to compare

# Set the width of the bars
bar_width = 0.35

# Create a list of test labels (e.g., Test 1, Test 2, Test 3)
test_labels = ["Stack", "Pyramid", "L"]

# Create an array of test indices for plotting
test_indices = np.arange(num_tests)

# Iterate through the tests and create a histogram for each
for test_idx in range(1):
    plt.figure(figsize=(8, 6), dpi=500)  # Adjust the figure size as needed
    plt.bar(test_indices - bar_width / 2, optimization_success_rates[test_idx, :], bar_width, label='Optimization')
    plt.bar(test_indices + bar_width / 2, objective_success_rates[test_idx, :], bar_width, label='Objective')

    # Set plot labels and title
    plt.ylabel('Success Rate')
    plt.title(f'Comparison: Success Rate')
    plt.xticks(test_indices, test_labels)
    plt.yticks([round(0.1*i,2) for i in range(11)], [round(0.1*i,2) for i in range(11)])
    plt.legend()
```
